```python
import pandas as pd

names =['timestamp', 'activityID', 'heartrate', 'IMUH_temp', 'IMUH_acc1', 'IMUH_acc2',
              'IMUH_acc3', 'IMUH_acc1_6g', 'IMUH_acc2_6g', 'IMUH_acc3_6g', 'IMUH_gyr1', 'IMUH_gyr2', 'IMUH_gyr3',
              'IMUH_mag1', 'IMUH_mag2', 'IMUH_mag3', 'IMUH_orientation1', 'IMUH_orientation2', 'IMUH_orientation3',
             'IMUH_orientation4', 'IMUC_temp', 'IMUC_acc1', 'IMUC_acc2', 'IMUC_acc3', 'IMUC_acc1_6g', 'IMUC_acc2_6g', 
             'IMUC_acc3_6g', 'IMUC_gyr1', 'IMUC_gyr2', 'IMUC_gyr3', 'IMUC_mag1', 'IMUC_mag2', 'IMUC_mag3', 
             'IMUC_orientation1', 'IMUC_orientation2', 'IMUC_orientation3', 'IMUC_orientation4', 'IMUA_temp', 
             'IMUA_acc1', 'IMUA_acc2', 'IMUA_acc3', 'IMUA_acc1_6g', 'IMUA_acc2_6g', 'IMUA_acc3_6g', 'IMUA_gyr1', 
             'IMUA_gyr2', 'IMUA_gyr3', 'IMUA_mag1', 'IMUA_mag2', 'IMUA_mag3', 'IMUA_orientation1', 'IMUA_orientation2', 
             'IMUA_orientation3', 'IMUA_orientation4', 'subjectID']
df = pd.DataFrame(columns=names)
for i in range(101,110) :

    df_temp = pd.read_csv('PAMAP2_Dataset/Protocol/subject'+ str(i)+'.dat', sep=" ",header=None, names=names)
    df_temp['subjectID'] = 'subject' + str(i)
    df = pd.concat([df, df_temp], ignore_index=True)
```
