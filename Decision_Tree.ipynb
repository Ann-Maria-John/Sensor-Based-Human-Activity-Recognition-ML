```python
#importing libraries

import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)

df = pd.read_csv('processed_data/HAR_cleaned.csv')
df.head()
```

```python
df.shape
```

```python
#implement PCA to reduce the number of features

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#separate the features from the target variable
X = df.drop(['activityID','subjectID','timestamp'], axis = 1)
y = df['activityID']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Assuming X and y are your features and labels
# Perform PCA

#standardize the features
scaler = StandardScaler()
#remove column names to avoid error
df_scaled = pd.DataFrame(scaler.fit_transform(X))



pca = PCA(n_components=10)
X_pca = pca.fit_transform(df_scaled)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Create a Decision Tree classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the model
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
