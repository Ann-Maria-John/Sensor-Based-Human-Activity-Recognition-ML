```python
#importing libraries

import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)

df = pd.read_csv('processed_data/HAR_cleaned.csv')
df.head()
```

```python
df.shape
```

```python
#implement PCA to reduce the number of features

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#separate the features from the target variable
X = df.drop(['activityID','subjectID','timestamp'], axis = 1)
y = df['activityID']
```

```python
scaler = StandardScaler()
#remove column names to avoid error
df_scaled = pd.DataFrame(scaler.fit_transform(X))

```

```python
#apply PCA
pca = PCA(n_components=10)
X_pca = pca.fit_transform(df_scaled)

df_pca = pd.DataFrame(pca.fit_transform(X_pca))
df_pca
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Assuming you have your features (X) and labels (y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Standardize the features (considered as regularization in this context)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the KNN classifier
n_neighbors = 5  # You can adjust this as part of regularization
knn_classifier = KNeighborsClassifier(n_neighbors=n_neighbors)

# Train the model
knn_classifier.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = knn_classifier.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Test Accuracy: {accuracy}')
```
