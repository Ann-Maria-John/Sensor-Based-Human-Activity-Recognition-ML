```python
def convert_seconds_to_datetime(seconds):
    specific_date = pd.Timestamp('2011-09-01')  # Specific date
    int_part = int(seconds)  # Extract integer part for seconds
    decimal_part = seconds - int_part  # Extract decimal part for milliseconds
    milliseconds = int(decimal_part * 1000)  # Convert decimal part to milliseconds
    return specific_date + pd.to_timedelta(int_part, unit='s') + pd.to_timedelta(milliseconds, unit='ms')

# Apply the conversion function to the 'timestamp' column
df1['timestamp'] = df1['timestamp'].apply(convert_seconds_to_datetime)
```

```python
df1.set_index('timestamp', inplace=True)
```

```python
df1.head()
```

```python
df1.shape
```

```python
import pandas as pd

def interpolate_missing_values(df, method='linear', window_size=None, alpha=None):
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            if method == 'linear':
                df[column].interpolate(method='linear', inplace=True)
            elif method == 'sma' and window_size is not None:
                df[column].fillna(df[column].rolling(window=window_size, min_periods=1).mean(), inplace=True)
            elif method == 'ema' and alpha is not None:
                df[column].fillna(df[column].ewm(alpha=alpha, min_periods=1).mean(), inplace=True)
    
    return df
```

```python
interpolated_df = interpolate_missing_values(df1, method='sma', window_size=3)
```

#### FFT Analysis

```python
from scipy.fft import fft, ifft
from scipy.signal import get_window
```

```python
# Selecting only the sensor columns
sensor_columns = merged_df.iloc[:, 3:33]

# Apply windowing to the sensor columns
window = get_window('hamming', len(sensor_columns))
windowed_data = sensor_columns * window[:, None]  # Apply window to each column

# Apply FFT on the windowed sensor columns
windowed_fft = np.fft.fft(windowed_data, axis=0)
frequencies = np.fft.fftfreq(len(sensor_columns), d=0.01)  # Assuming data is sampled every 0.01s

# Define a cutoff frequency for the low-pass filter
cutoff_frequency = 10  # Modify this as needed (in Hz)

# Find the indices of frequencies above the cutoff
cutoff_idx = np.abs(frequencies) > cutoff_frequency

# Apply the low-pass filter (set high-frequency components to zero)
windowed_fft_filtered = windowed_fft.copy()
windowed_fft_filtered[cutoff_idx, :] = 0

# Perform inverse FFT on the filtered data
filtered_ifft = np.fft.ifft(windowed_fft_filtered, axis=0)

# Create a new DataFrame for the denoised data
denoised_data = pd.DataFrame(filtered_ifft, columns=sensor_columns.columns)

# Plot original and denoised data for a specific sensor column (e.g., first column)
plt.figure(figsize=(10, 6))
plt.plot(sensor_columns.iloc[:, 0], label='Original Data')
plt.plot(denoised_data.iloc[:, 0], label='Denoised Data')
plt.title('Denoising Sensor Data with Windowing')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# Selecting only the sensor columns
sensor_columns = merged_df.iloc[:, 3:33]

# Apply windowing to the sensor columns
window = get_window('hamming', len(sensor_columns))
windowed_data = sensor_columns * window[:, None]  # Apply window to each column

# Calculate spectrogram for the noisy data
fs = 100  # Sampling frequency (Hz)
nperseg = 256  # Number of data points in each segment
noverlap = 128  # Number of overlapping points between segments

plt.figure(figsize=(10, 6))
plt.specgram(windowed_data.values.flatten(), Fs=fs, NFFT=nperseg, noverlap=noverlap, cmap='viridis')
plt.title('Spectrogram of Noisy Data')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.colorbar(label='Intensity')
plt.show()
```
