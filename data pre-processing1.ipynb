```python
#importing libraries

import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)
```

#### Subject 101

```python
#load data

with open('subject101.dat', mode='r', encoding='utf-8-sig') as admins:
    data1 = admins.read()
```

```python
#split string into list where each line is list item

data_list1 = data1.splitlines()
```

```python
# Split the list items by commas and create a list of lists

list_of_lists1 = [item.split() for item in data_list1]
```

```python
df1 = pd.DataFrame(list_of_lists1)
```

```python
df1.insert(loc = 0, column = 'subjectID', value = 101)
```

```python
df1.shape
```

```python
df1.head()
```

```python
df1.tail()
```
